# Querying a DataFrama

```python
import pandas as pd

df = pd.read_csv('Datasets/Admission_Predict.csv', index_col=0)
# let's clean the names
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
# now let's make a query to our dataset
admit_mask = df['chance of admit'] > 0.7
admit_mask
```

```python
# so to visulize the data related to earlier result
df.where(admit_mask).head()
```

```python
# to drop that NaN values
df.where(admit_mask).dropna().head()
```

```python
# Second example
df["gre score"].head()
```

```python
# or to put more columns
df[["gre score", "toefl score"]].head()
```

```python
# or to send a boolean mask
df[df["gre score"] > 320].head()
```

```python
# Now to combine multiple queries we use de & symbol
df[(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)]
# where is necessary to put the parentesis for each query
```

```python
# another way to do this no need of parentesis is
df[df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)]
```

# Manipulating DataFrame

```python
# let's start by bringing in pandas
import pandas as pd
df = pd.read_csv("Datasets/presidents.csv")
df.head()
```

```python
# so, let's begin cleaning the firstnames and lastnames
df["First"] = df['President']
df["First"] = df["First"].replace("[ ].*", "", regex=True)
df
```

```python
# there is another way to do that
del(df["First"]) #drop the column we made first
#now we define a function
def splitname(row):
    # let's extract the firstname
    row['First'] = row['President'].split(" ")[0]
    # and the lastname as well
    row['Last'] = row['President'].split(" ")[-1]
    # and we return the row
    return row
#to apply this function we make
df = df.apply(splitname, axis='columns')
df.head()
```

```python
# now let's drop the colums we made
del(df['First'])
del(df['Last'])
# and let's make once more time this but using regex
# out pattern to match the fist and lastname is
pattern = "(^[\w]*)(?:.* )([\w]*$)"
# and so we can extract the information we want to
df["President"].str.extract(pattern).head()
```

```python
# and to label each column we put name on the group regex
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names = df["President"].str.extract(pattern).head()
names
```

```python
# now we can just copy these into our main dataframe
df["First"] = names["First"]
df["Last"] = names["Last"]
df
```

```python
# Now let's clean up the Born column
df["Born"] = df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

```python
# we can modify the structure of this date, e.g.,
df["Born"] = pd.to_datetime(df["Born"])
df["Born"].head()
```
