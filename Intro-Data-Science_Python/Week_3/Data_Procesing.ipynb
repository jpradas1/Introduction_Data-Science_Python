# Merging DataFrames

```python
# In this episode we gonna use the pandas' merge function
import pandas as pd
# first let's create some dataframes
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
# and lets index these staff by name
staff_df = staff_df.set_index('Name')
# the another dataframe
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                         {'Name': 'Mike', 'School': 'Law'},
                         {'Name': 'Sally', 'School': 'Engineering'}])
# and lets index
student_df = student_df.set_index('Name')

print(staff_df.head())
print(student_df.head())
```

```python
# so to make the union of these, we gonna call merge() function
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
# where we input the df we want to make a union, 'outer'
# parameter means that we choose the union operation
# and the last two parameters indicate the joining columns
```

```python
# To make the intersection of these, instead of 'outer' we use 'inner' parameter
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
```

```python
# now if we want to see, for example, the staff information
# together with its school information as well we make a 
# right or left join
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
# becuase staff_df is the left dataframe we use left join
```

```python
# so as well as the previous, we can do right join
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
# in this case we see the student information and its related
# staff data
```

```python
# Also we can make this even when there's no index on df
# let's remove ourt index from both of our dataframes
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()

# and now lets make the early right join
pd.merge(staff_df, student_df, how='right', on='Name')
# in this case we use the paramater 'on' to choose the
# joining column
```

```python
# Now what happends when we have conflicts
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR', 
                          'Location': 'State Street'},
                         {'Name': 'Sally', 'Role': 'Course liasion', 
                          'Location': 'Washington Avenue'},
                         {'Name': 'James', 'Role': 'Grader', 
                          'Location': 'Washington Avenue'}])
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business', 
                            'Location': '1024 Billiard Avenue'},
                           {'Name': 'Mike', 'School': 'Law', 
                            'Location': 'Fraternity House #22'},
                           {'Name': 'Sally', 'School': 'Engineering', 
                            'Location': '512 Wilson Crescent'}])
# in this case, staff's location is the office address,
# but the student's location is the home address
# so when we make a join, pandas will create two differents
# columns to store both directions
pd.merge(staff_df, student_df, how='left', on='Name')
```

```python
# let's see now how works multi-index and multiple columns
staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 
                          'Role': 'Director of HR'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 
                          'Role': 'Course liasion'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 
                          'Role': 'Grader'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 
                            'School': 'Business'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 
                            'School': 'Law'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 
                            'School': 'Engineering'}])
# so depeding of which index we pass to pandas, there will be
# data that overlaping. In these cases we pass both index
pd.merge(staff_df, student_df, how='inner', on=['First Name', 'Last Name'])
# here we have applied union operation
```

```python
import numpy as np
import timeit

df = pd.read_csv('Datasets/census.csv')
df.head()
```

```python
# so let's clean the data
(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME', 'CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Bases 2010'}))
```

```python
# and we get a similar result by performing
df = df[df['SUMLEV']==50]
df.set_index(['STNAME', 'CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Bases 2010'})
```

```python
# so to know what method is faster we gonna use the timeit library
def first_approach():
    global df
    
    return (df.where(df['SUMLEV']==50)
              .dropna()
              .set_index(['STNAME', 'CTYNAME'])
              .rename(columns={'ESTIMATESBASE2010': 'Estimates Bases 2010'}))

df = pd.read_csv('Datasets/census.csv')

timeit.timeit(first_approach, number=10)
```

```python
# so the second method goes as
def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Bases 2010'})

df = pd.read_csv('Datasets/census.csv')

timeit.timeit(second_approach, number=10)
```

```python
# let's create a function that returns the maximun and minimun
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max':np.max(data)})
```

```python
df.apply(min_max, axis='columns').head()
```

```python
# again, there's another way to solve this
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row

df.apply(min_max, axis='columns')
```

```python
# and again, we can use lambdas to find the max values
rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012',
        'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']

df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
# let's treat whit another example
def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
# so, if we want sort data in fucntion of its direction
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
